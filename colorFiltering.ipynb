```python
from __future__ import print_function
import cv2 as cv
import argparse
import matplotlib.pyplot as plt
import os

max_value = 255
max_value_H = 255
low_H = 0
low_S = 0
low_V = 0
high_H = max_value_H
high_S = max_value
high_V = max_value
window_capture_name = 'Video Capture'
window_detection_name = 'Object Detection'
low_H_name = 'Low H'
low_S_name = 'Low S'
low_V_name = 'Low V'
high_H_name = 'High H'
high_S_name = 'High S'
high_V_name = 'High V'

def create_path(path):
    if not os.path.exists(os.path.dirname(path)):
        try:
            os.makedirs(os.path.dirname(path))
        except OSError as exc: # Guard against race condition
            if exc.errno != errno.EEXIST:
                raise
def classification(i): 
    frame = cv.imread("/home/ltruong/Desktop/hellboy/beet/data_beet/im_"+str(i)+".jpg")
    frame_HSV = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    frame_threshold = cv.inRange(frame_HSV, (low_H, low_S, low_V), (high_H, high_S, high_V))
    frame_threshold_beet = cv.inRange(frame_HSV, (40, 105, low_V), (high_H, high_S, high_V))
    frame_threshold_thistle = cv.inRange(frame_HSV, (40, 32, 108), (high_H, 80, 120))
    im3 = frame.copy()

    contours, hierarchy = cv.findContours(frame_threshold_beet,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for c in contours:
        rect = cv.boundingRect(c)
        x,y,w,h = rect
        cv.rectangle(im3,(x,y),(x+w,y+h),(0,255,0),2)

    contours1, hierarchy1 = cv.findContours(frame_threshold_thistle,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for c in contours1:
        rect = cv.boundingRect(c)
        x,y,w,h = rect
        cv.rectangle(im3,(x,y),(x+w,y+h),(255,0,0),2)

    create_path("/home/ltruong/Desktop/hellboy/beet/results/"+"img_"+str(i)+"/real.jpg")
    create_path("/home/ltruong/Desktop/hellboy/beet/results/"+"img_"+str(i)+"/detect.jpg")
    create_path("/home/ltruong/Desktop/hellboy/beet/results/"+"img_"+str(i)+"/beet.jpg")
    create_path("/home/ltruong/Desktop/hellboy/beet/results/"+"img_"+str(i)+"/thistle.jpg")
    
    
    cv.imwrite("/home/ltruong/Desktop/hellboy/beet/results/"+"img_"+str(i)+"/real.jpg",frame)
    cv.imwrite("/home/ltruong/Desktop/hellboy/beet/results/"+"img_"+str(i)+"/detect.jpg",im3)
    cv.imwrite("/home/ltruong/Desktop/hellboy/beet/results/"+"img_"+str(i)+"/beet.jpg",frame_threshold_beet)
    cv.imwrite("/home/ltruong/Desktop/hellboy/beet/results/"+"img_"+str(i)+"/thistle.jpg",frame_threshold_thistle)
    print("done")

```
